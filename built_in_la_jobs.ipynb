```python
from bs4 import BeautifulSoup, Tag
import urllib
```

```python
base = 'https://www.eventbrite.com/e/built-in-las-top-companies-hiring-tickets-42297906176'
html = urllib.request.urlopen(base)
soup = BeautifulSoup(html, 'html5lib')
soup
```

```python
import re
div = soup.find('div', attrs={'class':'js-xd-read-more-toggle-view read-more__toggle-view'})
links = div.find_all('a')
subpages = []
for a in links:
    subpage = {}
    if 'company' in a['href']:
        subpage['company'] = re.search('company/(?P<comp>.*)$',a['href']).group('comp')
        subpage['url'] = a['href']
        subpages.append(subpage)        
```

```python
for s in subpages:
    print(s['url'])
    sub_html = urllib.request.urlopen(s['url'])
    sub_soup = BeautifulSoup(sub_html, 'html5lib')
    jobs = sub_soup.find_all('a', href=re.compile('jobs/engineer'))
    for j in jobs:
        print(j.get_text())
```
